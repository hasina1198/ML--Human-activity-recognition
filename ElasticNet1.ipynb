```python
import pandas as pd
import numpy as np
from sklearn import model_selection
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix


Xtrain = pd.read_table('X_train.txt', delim_whitespace = True, header = None)
ytrain = pd.read_table('y_train.txt', delim_whitespace = True, header = None)
Xtest = pd.read_table('X_test.txt', delim_whitespace = True, header = None)
ytest = pd.read_table('y_test.txt', delim_whitespace = True, header = None)

print (Xtrain.shape)
print (ytrain.shape)
print (Xtest.shape)
print (ytest.shape)
```

```python
from sklearn.linear_model import SGDClassifier

alpha= [1e-4,3e-4,1e-3,3e-3, 1e-2,3e-2]
l1= [0,0.15,0.5,0.7,1]
f= []
for i in range(0,6):
    for j in range(0,5): 
        f.append(model_selection.cross_val_score(SGDClassifier(loss='log', penalty='elasticnet', alpha=alpha[i], l1_ratio=l1[j], max_iter=5), Xtrain, ytrain, cv=model_selection.KFold(n_splits = 10), scoring = 'f1_weighted').mean())         
n=list(range(0,30))
k = n[f.index(max(f))]+1
ab = np.ceil(k/6)-1
cd = np.remainder(k,5)
if (cd != 0):
    cd-=1 
else:
    cd = 4
print("alpha val: ",alpha[int(ab)])
print("l1 val: ",l1[int (cd)])
```

```python
from mpl_toolkits.mplot3d import Axes3D

enet = SGDClassifier(loss='log', penalty='elasticnet', alpha=0.001, l1_ratio=0.5, max_iter=5) 
enet.fit(Xtrain,ytrain)

accuracy = metrics.accuracy_score(ytest, enet.predict(Xtest))
f1 = metrics.f1_score(ytest, enet.predict(Xtest), average = 'weighted')
confusion_matrix = metrics.confusion_matrix(ytest, enet.predict(Xtest))
print ("Accuracy: ",accuracy)
print ("F1 Score: ",f1)
print ("Confusion Matrix: ",confusion_matrix)
```
